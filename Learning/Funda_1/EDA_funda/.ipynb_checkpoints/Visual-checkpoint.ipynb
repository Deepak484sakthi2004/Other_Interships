```python
import matplotlib.pyplot as plt

%matplotlib inline # it is an alias , that is used to not always code plt.show()
```

```python
import numpy as np
```

```python
arr1=np.linspace(0,50,20)
arr2=np.linspace(40,50,20)
print(arr1,arr2)
```

```python

#SCATTERPLOT NUM VS NUM

plt.scatter(arr1,arr2,c='black')
#press shift+tab i=to know more 
#s is the solid value(in pixels)

```

```python
plt.xlabel('arr1')
plt.ylabel('arr2')
plt.title('Bi-variate')
```

```python
#LINEPLOT NUM VS NUM

plt.plot(arr1,arr2)
```

```python
plt.plot(arr1,arr1,'bo')
```

```python

plt.plot(arr1,arr1,'bo--')
```

```python
plt.plot(arr1,arr2,'g*--',linestyle='dashed',linewidth=2,markersize=12)
```

 SUBPLOTS

```python
plt.subplot(2,2,4)
plt.plot(arr1,arr1,'ro--',markersize=2.9)
plt.title('plot4')


plt.subplot(2,2,1)
plt.plot(arr1,arr1,'g--')


#plt.subplot(rows,columns,position)
```

# SEABORN

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

DATA ANALYSIS FROM WHAT I KNOW

```python
print('maximum tip:',df['tip'].max())
print('avg tip:',df['tip'].mean())
print('avg of no outlier tip',df['tip'].median())
```

```python
df.plot(kind='scatter',x='total_bill',y='tip')
```

```python

df['tip'].plot(kind='kde')
```

```python
df['tip'].plot(kind='box',figsize=(5,10))
```

```python
df_1=df[df['tip']<=40]
df_1.plot(kind='kde',figsize=(10,5))
```

```python
df_1.boxplot(by='sex',column='tip')
```

```python
df.corr()
#only the numerical features 
```

```python
sns.heatmap(df.corr())
```
