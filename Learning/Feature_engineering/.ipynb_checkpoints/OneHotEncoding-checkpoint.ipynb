here it is not possible to select the column for feature engineering, as we have to create new columns for every unique features.So the better way is , select the top 10 most recurring attributes and make the rest of the attributes to 0 , i.e..fit those allin only one column...

```python
top10=df['X0'].value_counts().sort_values(ascending=False)[:15]

print(top10[4]) # here the index of every array is the top10 column attributes and the element 
                # in the position is the count of their recurrence
top10
```

and perform the OHE on these top 10 attributes!!

```python
pd.get_dummies(df,drop_first='True').shape
```

I have got 180+ new columns that seems to be quite hard!!

```python
top15=[for x in df['X0'].value_counts().index]
top15
```

```python
# i thought it should be the way , but x has to come at front 


#here comes the 15 most frenqent attributes!!
top15=[x for x in df['X0'].value_counts().sort_values(ascending=False).head(15).index]
top15
```

```python
# the encoding!!
for i in top15:
    df[i]=np.where(df['X0']==i,1,0)
df.info()
```

```python
df[['X0']+top15].head()
```

```python
# Function OHE
def OHE(df,var,top15):
    for i in top15:
        df[var+'_'+i]=np.where(df[var]==i,1,0)
    
```

```python
top15=[x for x in df['X1'].value_counts().head(15).index]

OHE(df,'X1',top15)
df.info()
```

# ENCODING

Count of Frequency

```python
df=pd.read_csv('OHE.csv')
df=df.iloc[:,:9]
df
```

```python
for i in df.columns:
    print(i,':',len(df[i].unique()))
```

```python
# so in Count of Frequency, we create a dict with the recurring value and the count of it

df['X0'].value_counts().to_dict()
```

```python
df_freq_d=df['X0'].value_counts().to_dict()
```

```python
df['X0'].head(10)
```

```python
#map the dict with the respective feature column

df['X0']=df['X0'].map(df_freq_d)
```

```python
df['X0']
```

```python
df
```

The main disadvantage of this Count of frequency is that if any other feature column as the same counts
then the uniqueness of the data will be lost

```python
#function for Count of Frequency
def COF(df,var):
    df_freq=df[var].value_counts().to_dict()
    
    df[var]=df[var].map(df_freq)
    
    
```

```python
COF(df,'X2')
```
