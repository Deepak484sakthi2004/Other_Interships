CODE FOR WEB SCRAPPING


```python
title = []
rating = []
price = []
review=[]

for i in range(1, 39):
    URL='https://www.flipkart.com/search?q=hp+printers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    
    
    for x in soup.find_all('div', attrs={'class' : '_4ddWXP'}):
        
        div_tags = soup.find_all('div', {'class': 'CXW8mj'})
        for div_tag in div_tags:
            image_tag = div_tag.find('img')
            text = image_tag['alt']
            title.append(text)

        mrp = x.find('div', attrs={'class' : '_30jeq3'})
        if mrp is None:
            price.append(np.NaN)
        else:
            price.append(mrp.text)

        rate = x.find('div', attrs={'class' : '_3LWZlK'})
        if rate is None:
            rating.append(np.NaN)
        else:
            rating.append(rate.text)
               
    review_t = soup.find_all("div", class_="_2tzWk4")
#print(review)

    for tag in review_t:
        review.append(tag.text)
```

```python
print(len(title))
print(len(price))
print(len(rating))
print(len(review))
```

```python
rating = np.array(rating)
title = np.array(title)
price= np.array(price)
review=np.array(review)
```

```python
max_length = max(len(rating), len(title), len(price),len(review))
p_review = np.pad(review, (0, max_length-len(review)), mode='constant', constant_values=0)
p_rating = np.pad(rating, (0, max_length-len(rating)), mode='constant', constant_values=0)
p_price = np.pad(price, (0, max_length-len(price)), mode='constant', constant_values=0)
```

```python
df = pd.DataFrame({'Product' : title, 'Rating' : p_rating, 'MRP' : p_price,'Review':p_review})
```

```python


# save the dataframe to an Excel file
df.to_excel('data.xlsx', index=True)
```
