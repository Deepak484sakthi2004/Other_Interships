```python
#importing modules
import pandas as pd
import numpy as np

#dataset
df=pd.read_csv('monthly_.csv')
print(df.head())
df.tail()
```

```python
df.info()
```

```python
#conventing the Date from Obj to date format
df['Date']=pd.to_datetime(df['Date'])
```

```python
# year and month columns
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
#dropping Date
df=df.drop('Date',1)
```

```python
df.info()
```

## EDA

```python
#Univariate on Price
df['Price'].plot(kind='kde')
```

```python
#Bivariate on Year verses Price

df.plot(kind='scatter',x='Year',y='Price',figsize=(20,10))
```
